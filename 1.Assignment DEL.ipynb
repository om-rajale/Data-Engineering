```python
import pandas as pd
import re
import numpy as np
```

```python
df = df[0].str.split(',', expand=True)
df.columns = ['emp_id', 'name', 'email', 'joining_date', 'department', 'salary']
```

```python
df
```

```python
df.drop_duplicates()
```

```python
df['name'] = df['name'].fillna('Om')
```

```python
df
```

```python
def is_valid_email(email):
    if pd.isna(email):
        return False
    return re.match(r"[^@]+@[^@]+\.[^@]+", str(email)) is not None

df['email'] = df['email'].apply(lambda x: x if is_valid_email(x) else np.nan)
df
```

```python
df['email'] = df['email'].fillna('abc@gmail.com')
```

```python
df
```

```python
df.to_csv('customers_cleaned2.csv', index=False)
```

```python
df.drop(0, inplace=True)
df
```

```python
df['name'] = df['name'].fillna('Om')
df
```

```python
# Force non-numeric to NaN
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')

# Remove NaN values and keep only positive salaries
valid_salary = df['salary'].dropna()
valid_salary = valid_salary[valid_salary > 0]

# Median salary from valid values
median_salary = valid_salary.median()

# Replace NaN or non-positive salaries with the median
df['salary'] = df['salary'].apply(
    lambda x: median_salary if pd.isna(x) or x <= 0 else x
)

df
```

```python
import numpy as np

# Normalize the name column
df['name'] = df['name'].astype(str).str.strip()
df['name'] = df['name'].replace(['NaN'], np.nan)
df['name'] = df['name'].fillna('Om')
df
```

```python
df['joining_date'] = df['joining_date'].fillna(pd.Timestamp('2025-08-11'))
df
```

```python
df['joining_date'] = pd.to_datetime(df['joining_date'].replace(r'^\s*$', np.nan, regex=True), errors='coerce').fillna('2025-08-11')
df
```
